```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df_clothing = pd.read_csv('../Data/Fashion_Dataset.csv')
```

```python
# Inspect the dataset
print(df_clothing.head())
print(df_clothing.tail())

# What is the shape of the data
print('The shape of the dataset is: ', df_clothing.shape)
```

```python
# Summary statistics and missing values
summary = df_clothing.describe(include='all')
print("Statistical summary of dataset: ", summary)
missing_values = df_clothing.isnull().sum()
print(missing_values)

# Columns in dataset
print(df_clothing.columns)
```

```python
# Visualization

# Plot for price and ratings
plt.figure(figsize=(10, 6))
plt.scatter(df_clothing['price'], df_clothing['avg_rating'], alpha=0.5)
plt.title('Price v/s Rating')
plt.ylabel('Average Rating')
plt.xlabel('Price')
plt.grid(True)
plt.show()

# Price Distribution Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='price', data=df_clothing)
plt.title('Price Distribution of Products')
plt.xlabel('Price')
plt.show()


# Count Plot of Brands
plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=df_clothing, palette="plasma")
plt.title('Distribution of Product Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()

# Correlation Matrix for Numerical Attributes
correlation_matrix = df_clothing[['price', 'ratingCount', 'avg_rating']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Numerical Attributes')
plt.show()

# Displaying correlation matrix and dataset statistics
print("Correlation Matrix:\n", correlation_matrix)
```
